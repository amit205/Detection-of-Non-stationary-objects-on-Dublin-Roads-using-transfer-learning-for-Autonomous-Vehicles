```python
import matplotlib.pyplot as plt
import pandas as pd
train = pd.read_csv('train_labels.csv')
test = pd.read_csv('test_labels.csv')
```

```python
train.count()
```

```python
train.head()
```

```python
test.head()
```

```python
group_train = train.groupby('class')
group_test = test.groupby('class')
```

```python
plot_train = group_train.count()
plot_train
```

```python
y = plot_train['filename']
x = plot_train.index.values
f, ax = plt.subplots(figsize=(18,8))
plt.bar(x, y, alpha=1, color = 'red', label = 'train')
plt.xticks(x)
plt.ylabel('Count')
plt.title('Objects')
y1 = plot_test['filename']
x1 = plot_test.index.values
plt.bar(x1, y1, align='center', alpha=1, label='test')
ax.legend(fontsize=14)
```

```python
plot_test = group_test.count()
plot_test
```

```python
f, ax = plt.subplots(figsize=(15,8))
y1 = plot_test['filename']
x1 = plot_test.index.values
plt.bar(x1, y1, align='center', alpha=1)
plt.xticks(x1)
plt.ylabel('Count')
plt.title('Objects')
plt.show()
```
